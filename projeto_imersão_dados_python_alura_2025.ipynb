```python
# Legenda de Senioridades SE = Senior | MI = Pleno | EN = Junior | EX = Executivo(CEO Level)
senioridade_traducao = {
    'SE': 'Senior',
    'MI': 'Pleno',
    'EN': 'Junior',
    'EX': 'Executivo'
}

df['senioridade'] = df['senioridade'].replace(senioridade_traducao)
#df['senioridade'] = df['senioridade'].map(senioridade_traducao)
# REPLACE - foi usado para substituir valores específicos na coluna 'senioridade'.
#   Ele procurou pelas abreviações ('SE', 'MI', etc.) e trocou cada uma pelo seu valor correspondente
#   no dicionário senioridade_traducao ('Senior', 'Pleno', etc.). É bom para trocar alguns valores por outros.
# MAP - (nesse contexto específico) também poderia ter sido usado para fazer a mesma troca,
#   aplicando o dicionário a cada item da coluna. A principal diferença é que map é mais focado em transformar cada elemento
#   de uma série com base em um mapeamento, e se um valor não for encontrado no dicionário, ele colocaria NaN (valor ausente)
#   por padrão, a menos que você especifique o contrário. replace é mais flexível para diferentes tipos de substituição.

df["senioridade"].value_counts()
```

```python
# Tipo de Contrato de Trabalho ["tipo_emprego"]
# FT = Full Time (Horário Integral, comercial aqui no Brasil)
# CT = Contato Trabalho por determinado período de tempo
# PT = Parcial Time (Meio Período)
# FL = FreeLance (Trabalho por projetos ou em deteminadas atividades)
contrato_traducao = {
    'FT': 'Integral',
    'CT': 'Contrato',
    'PT': 'Parcial',
    'FL': 'Freelance'
}

df['contrato'] = df['contrato'].replace(contrato_traducao)

df["contrato"].value_counts()
```

```python
# Legenda/Total tipo de trabalho 0 = Presencial | 100 = Remoto | 50 = Híbrido
remoto_traducao = {
    0: 'Presencial',
    100: 'Remoto',
    50: 'Hibrido'
}

df['remoto'] = df['remoto'].replace(remoto_traducao)

df["remoto"].value_counts()
```

```python
# Legenda Tamanho Empresa M = Medium (Médio) | L = Large (Grande) | S = Small (Pequeno)
tamanho_empresa_traducao = {
    'M': 'Media',
    'L': 'Grande',
    'S': 'Pequena'
}

df['tamanho_empresa'] = df['tamanho_empresa'].replace(tamanho_empresa_traducao)

df["tamanho_empresa"].value_counts()
```

```python
import numpy as np

df_salarios = pd.DataFrame({
    'nome': ["Ana", "Bruno", "Carlos", "Daniele", "val"],
    'salario': [4000, np.nan, 5000, np.nan, 100000]
})

# Médias Salariais para:

# Média - é o valor central (soma/quantidade)
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))
#display(df_salarios)

# Mediana - é o valor do meio em dados ordenados
df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())#.round(2))
#display(df_salarios)

# Moda - é o valor que mais se repete - (foi pego o primeiro valor pois pode retornar múltiplos valores)
df_salarios['salario_moda'] = df_salarios['salario'].fillna(df_salarios['salario'].mode()[0].round(2))
display(df_salarios)
```

```python
import pycountry

# Função para converter ISO-2 para ISO-3
def iso2_to_iso3(code):
    try:
        return pycountry.countries.get(alpha_2=code).alpha_3
    except:
        return None

# Criar nova coluna com código ISO-3 conforme a sigla do pais
df_limpo['residencia_iso3'] = df_limpo['residencia'].apply(iso2_to_iso3)

# Calcular média salarial por país (ISO-3)
df_ds = df_limpo[df_limpo['cargo'] == 'Data Scientist']
media_ds_pais = df_ds.groupby('residencia_iso3')['usd'].mean().reset_index()

# Gerar o mapa
fig = px.choropleth(media_ds_pais,
                    locations='residencia_iso3',
                    color='usd',
                    color_continuous_scale='rdylgn',
                    title='Salário médio de Cientista de Dados por país',
                    labels={'usd': 'Salário médio (USD)', 'residencia_iso3': 'País'})

fig.show()
```
